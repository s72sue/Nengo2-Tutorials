# Routed Sequencing with Cleanup Memory

The previous model on Routed Sequencing used the basal ganglia model to cycle through a 5 element sequence, where an arbitrary start was presented to the model. 

In that model, the network was specified using abstract rules. However, it is often useful to create ensembles of neurons so that their parameters can be directly adjusted. This demo thus extends the routed sequencing demo by creating a scalar ensemble 'cleanupA' and then projecting the state of the cortex on to this ensemble using a transformation matrix 'pd', specified by the vocabulary vector 'A'. This implies that each component of the state vector (i.e., the cortex) is multiplied by the corresponding component of the 'A' vector and summed to produce a one dimensional quantity represented by the clean-up ensemble. This operation is similar to the dot product of the state vector and the vocabulary vector 'A'.

```python
# Setup for the notebook
import matplotlib.pyplot as plt
%matplotlib inline

import nengo
from nengo import spa
from nengo.spa import Vocabulary
```

## Create the Model

The model is the same as the routed sequencing model with an additional transformation matrix 'pd' specified by the vocabulary vector 'A' and a cleanup ensemble (cleanupA).


```python
# Number of dimensions for the Semantic Pointers
dimensions = 16

# Make a model object with the SPA network
model = spa.SPA(label='Routed_Sequence with cleanupA')

with model:
    # Specify the modules to be used
    model.cortex = spa.Buffer(dimensions=dimensions)
    model.vision = spa.Buffer(dimensions=dimensions) 
    # Specify the action mapping
    actions = spa.Actions(
        'dot(vision, START) --> cortex = vision',
        'dot(cortex, A) --> cortex = B',
        'dot(cortex, B) --> cortex = C',
        'dot(cortex, C) --> cortex = D',
        'dot(cortex, D) --> cortex = E',
        'dot(cortex, E) --> cortex = A'
    )
    model.bg = spa.BasalGanglia(actions=actions)
    model.thal = spa.Thalamus(model.bg)
    
    # Change the seed of this RNG to change the vocabulary
    rng = np.random.RandomState(0)
    vocab = Vocabulary(dimensions=dimensions)

    # Create the transformation matrix (pd) and the cleanup ensemble (cleanupA) 
    pd = [vocab['A'].v.tolist()] 
    cleanupA = nengo.Ensemble(100, dimensions=1)
    
    # Function that provides the model with an initial input semantic pointer.
    def start(t):
        if t < 0.05:
            return '0.8*START+D'
        else:
            return '0'

    # Input
    model.input = spa.Input(vision=start)
    
    # Projecting the state of the cortex on to the cleanup ensemble using a transformation matrix 'pd'.
    nengo.Connection(model.cortex.state.output, cleanupA, transform=pd)
```

## Add Probes to Collect Data

```python
with model:
    cortex = nengo.Probe(model.cortex.state.output, synapse=0.01)
    vision = nengo.Probe(model.vision.state.output, synapse=0.01)
    actions = nengo.Probe(model.thal.actions.output, synapse=0.01)
    utility = nengo.Probe(model.bg.input, synapse=0.01)
    cleanup_probe = nengo.Probe(cleanupA, synapse=0.01)
```

## Run the Model

```python
# Create the simulator object
sim = nengo.Simulator(model)
# Simulate the model for 0.5 seconds
sim.run(0.5)
```

## Plot the Results

```python
fig = plt.figure(figsize=(12,8))
p1 = fig.add_subplot(4,1,1)
p1.plot(sim.trange(), model.similarity(sim.data, vision))
p1.legend(model.get_output_vocab('vision').keys, fontsize='x-small')
p1.set_ylabel('Vision')

p2 = fig.add_subplot(4,1,2)
p2.plot(sim.trange(), model.similarity(sim.data, cortex))
p2.legend(model.get_output_vocab('cortex').keys, fontsize='x-small')
p2.set_ylabel('Cortex')

p3 = fig.add_subplot(4,1,3)
p3.plot(sim.trange(), sim.data[cleanup_probe], 'k')
p3.set_ylabel('CleanupA')
p3.set_ylim(-0.6,0.8)

fig = plt.figure(figsize=(12,8))
p4 = fig.add_subplot(4,1,1)
p4.plot(sim.trange(), sim.data[actions])
p4.set_ylabel('Action')

p5 = fig.add_subplot(4,1,2)
p5.plot(sim.trange(), sim.data[utility])
p5.set_ylabel('Utility')


fig.subplots_adjust(hspace=0.2)
```

Since the cleanup operation is similar to a dot product between the state of the cortex (state vector) and the defined vocabulary vector 'A', the value of the cleanupA population should rise only when the cortex is in state 'A'. The response of the cleanup population is shown in the third plot above.
