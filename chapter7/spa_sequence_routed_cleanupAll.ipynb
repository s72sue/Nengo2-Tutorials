# Routed Sequencing with Cleanup all Memory

The previous demo on Routed Sequencing with cleanup memory used a scalar ensemble to project only state 'A' of the cortex. In this demo, we will create project all the states of cortex onto an ensemble of neurons as the cortex cycles through a five element sequence.

This demo thus extends the routed sequencing demo by creating a scalar ensemble 'cleanup' and then projecting all the states of the cortex on to this ensemble using a transformation matrix 'pd_new', specified by the vectors in the vocabulary. 

```python
# Setup for the notebook
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np 

import nengo
from nengo import spa
from nengo.spa import Vocabulary
```

## Step 1: Create the model

The model is the same as the routed sequencing model with an additional transformation matrix 'pd_new' specified by the vocabulary vectors and a cleanup ensemble.


```python
# Number of dimensions for the Semantic Pointers
dimensions = 16

# Make a model object with the SPA network
model = spa.SPA(label='Routed_Sequence with cleanupAll')

with model:
    # Specify the modules to be used
    model.cortex = spa.Buffer(dimensions=dimensions)
    model.vision = spa.Buffer(dimensions=dimensions) 
    # Specify the action mapping
    actions = spa.Actions(
        'dot(vision, START) --> cortex = vision',
        'dot(cortex, A) --> cortex = B',
        'dot(cortex, B) --> cortex = C',
        'dot(cortex, C) --> cortex = D',
        'dot(cortex, D) --> cortex = E',
        'dot(cortex, E) --> cortex = A'
    )
    model.bg = spa.BasalGanglia(actions=actions)
    model.thal = spa.Thalamus(model.bg)
    
    # Change the seed of this RNG to change the vocabulary
    rng = np.random.RandomState(0)
    vocab = Vocabulary(dimensions=dimensions, rng=rng)


    # Create the transformation matrix (pd_new) and the cleanup ensemble (cleanup)      
    vsize = len((model.get_output_vocab('cortex').keys)) - 1

    pd = []
    for item in range(vsize):
        pd.append(model.get_output_vocab('cortex').keys[item+1])  
    
    pd_new = []
    for element in range(vsize):
        pd_new.append([vocab[pd[element]].v.tolist()]) 
      
    cleanup = nengo.Ensemble(300, dimensions=vsize)
```

## Step 2: Provide input to the model

Specify a function that provides the model with an initial input semantic pointer.

```python
def start(t):
    if t < 0.05:
        return '0.8*START+D'
    else:
        return '0'

with model:
    model.input = spa.Input(vision=start)
```

## Step 3: Connect the Model Elements

Project the states of the cortex on to the cleanup ensemble using a transformation matrix 'pd'.

```python
with model:
    for i in range(5):
        nengo.Connection(model.cortex.state.output, cleanup[i], transform=pd_new[i])
```

## Step 4: Probe the output

```python
with model:
    cortex = nengo.Probe(model.cortex.state.output, synapse=0.01)
    vision = nengo.Probe(model.vision.state.output, synapse=0.01)
    actions = nengo.Probe(model.thal.actions.output, synapse=0.01)
    utility = nengo.Probe(model.bg.input, synapse=0.01)
    cleanup_probe = nengo.Probe(cleanup, synapse=0.01)
```

## Step 5: Run the model

```python
# Create the simulator object
sim = nengo.Simulator(model)
# Simulate the model for 0.5 seconds
sim.run(0.5)
```

## Step 6: Plot the results

```python
fig = plt.figure(figsize=(12,8))
p1 = fig.add_subplot(4,1,1)
p1.plot(sim.trange(), model.similarity(sim.data, vision))
p1.legend(model.get_output_vocab('vision').keys, fontsize='x-small')
p1.set_ylabel('Vision')

p2 = fig.add_subplot(4,1,2)
p2.plot(sim.trange(), model.similarity(sim.data, cortex))
p2.legend(model.get_output_vocab('cortex').keys, fontsize='x-small')
p2.set_ylabel('Cortex')

p3 = fig.add_subplot(4,1,3)
p3.plot(sim.trange(), sim.data[cleanup_probe], 'k')
p3.set_ylabel('CleanupA')
p3.set_ylim(-0.6,0.8)

fig = plt.figure(figsize=(12,8))
p4 = fig.add_subplot(4,1,1)
p4.plot(sim.trange(), sim.data[actions])
p4.set_ylabel('Action')

p5 = fig.add_subplot(4,1,2)
p5.plot(sim.trange(), sim.data[utility])
p5.set_ylabel('Utility')


fig.subplots_adjust(hspace=0.2)
```

Since the cleanup operation is similar to a dot product between the states of the cortex and the defined vocabulary vectors, the value of the cleanup population in a particular dimension should rise only when the cortex is in a state corresponding to that dimension. The response of the cleanup population is shown in the third plot above.
